```python
import pandas as pd
```

```python
meteorData = pd.read_csv("Meteorite_Landings.csv")
meteorData.head()
```

```python
meteorDataClean = meteorData.drop(columns=["States", "Counties", "fall", "nametype", "id"]).dropna()
meteorDataClean.head()
```

```python
meteorDataClean['year'] = meteorDataClean['year'].str.split(" ").str[0]
```

```python
meteorDataClean['year'] = meteorDataClean['year'].str.split("/").str[-1]
```

```python
meteorDataClean = meteorDataClean.reset_index()
meteorDataClean = meteorDataClean.rename(columns={"index": "id", "mass (g)": "mass_grams"})
meteorDataClean = meteorDataClean.set_index("id")
```

```python
meteorDataClean.head()
```

```python
meteorDataClean.to_csv("Clean_Meteor_Data.csv")
```

```python
meteorDataClean.to_json("Clean_Meteor_Data.json")
```

```python
meteorClassData = pd.DataFrame(meteorDataClean.groupby(by="recclass").count()["year"])
meteorClassData = meteorClassData.rename(columns={"year": "count_of_class"})
meteorClassData.head()
meteorClassData.to_csv("test.csv")
```

```python
meteorClassDataGrouped = meteorClassData.reset_index()
meteorClassDataGrouped['class'] = meteorClassDataGrouped['recclass'].str.extract('('
                                         'Acapulcoite|'
                                         'Achondrite|'
                                         'Angrite|'
                                         'Aubrite|'
                                         'Brachinite|'
                                         'Chondrite|'
                                         'C|'
                                         'Diogenite|'
                                         'Eucrite|'
                                         'E|'
                                         'Howardite|'
                                         'H|'
                                         'Iron, IAB|'
                                         'Iron, IIE|'
                                         'Iron, I|'
                                         'Iron|'
                                         'K|'
                                         'Lodranite|'
                                         'Lunar|'
                                         'L|'
                                         'Martian|'
                                         'Mesosiderite|'
                                         'OC|'
                                         'Pallasite|'
                                         'Relict iron|'
                                         'Relict OC|'
                                         'R|'
                                         'Stone|'
                                         'Ureilite|'
                                         'Winonaite'
                                         ')', expand=True).dropna()
meteorClassDataGrouped['class'] = meteorClassDataGrouped['class'].replace({
    'Stone': 'Ungrouped Stony Meteorites',
    'Chondrite': 'Ungrouped Chondrites',
    'C': 'Carbonaceous Chondrites',
    'OC': 'Ordinary Chondrites',
    'Relict OC': 'Ordinary Chondrites',
    'H': 'Ordinary Chondrites',
    'L': 'Ordinary Chondrites',
    'E': 'Enstatite Chondrites',
    'R': 'Minor Chondrites',
    'K': 'Minor Chondrites',
    'Achondrite': 'Ungrouped Achondrites',
    'Acapulcoite': 'Primitive Achondrites',
    'Lodranite': 'Primitive Achondrites',
    'Winonaite': 'Primitive Achondrites',
    'Howardite': 'Asteroidal Achondrites',
    'Eucrite': 'Asteroidal Achondrites',
    'Diogenite': 'Asteroidal Achondrites',
    'Angrite': 'Asteroidal Achondrites',
    'Aubrite': 'Asteroidal Achondrites',
    'Ureilite': 'Asteroidal Achondrites',
    'Brachinite': 'Asteroidal Achondrites',
    'Lunar': 'Lunar Meteorites',
    'Martian': 'Martian Meteorites',
    'Pallasite': 'Pallasites',
    'Mesosiderite': 'Mesosiderites',
    'Iron': 'Ungrouped Iron Meteorites',
    'Relict iron': 'Ungrouped Iron Meteorites',
    'Iron, IAB': 'Non-Magmatic Iron Meteorites',
    'Iron, IIE': 'Non-Magmatic Iron Meteorites',
    'Iron, I': 'Magmatic Iron Meteorites'
})
meteorClassDataGrouped = pd.DataFrame(meteorClassDataGrouped.groupby(by='class').sum())
meteorClassDataGrouped = meteorClassDataGrouped.reset_index()
meteorClassDataTier3 = meteorClassDataGrouped.rename(columns={"class": 'class_tier_3'})
meteorClassDataTier3
```

```python
def class_tier_2(row):
    if row['class_tier_3'] == 'Minor Chondrites':
        return 'Chondrites'
    elif row['class_tier_3'] == 'Carbonaceous Chondrites':
        return 'Chondrites'
    elif row['class_tier_3'] == 'Enstatite Chondrites':
        return 'Chondrites'
    elif row['class_tier_3'] == 'Ordinary Chondrites':
        return 'Chondrites'
    elif row['class_tier_3'] == 'Ungrouped Chondrites':
        return 'Chondrites'
    elif row['class_tier_3'] == 'Primitive Achondrites':
        return 'Achondrites'
    elif row['class_tier_3'] == 'Asteroidal Achondrites':
        return 'Achondrites'
    elif row['class_tier_3'] == 'Lunar Meteorites':
        return 'Achondrites'
    elif row['class_tier_3'] == 'Martian Meteorites':
        return 'Achondrites'
    elif row['class_tier_3'] == 'Ungrouped Achondrites':
        return 'Chondrites'
    else:
        return row['class_tier_3']
    
meteorClassDataTier2 = meteorClassDataTier3
meteorClassDataTier2['class_tier_2'] = meteorClassDataTier2.apply(class_tier_2, axis=1)
meteorClassDataTier2
```

```python
def class_tier_1(row):
    if row['class_tier_2'] == 'Chondrites':
        return 'Stony Meteorites'
    elif row['class_tier_2'] == 'Achondrites':
        return 'Stony Meteorites'
    elif row['class_tier_2'] == 'Ungrouped Stony Meteorites':
        return 'Stony Meteorites'
    elif row['class_tier_2'] == 'Mesosiderites':
        return 'Stony-Iron Meteorites'
    elif row['class_tier_2'] == 'Pallasites':
        return 'Stony-Iron Meteorites'
    elif row['class_tier_2'] == 'Magmatic Iron Meteorites':
        return 'Iron Meteorites'
    elif row['class_tier_2'] == 'Non-Magmatic Iron Meteorites':
        return 'Iron Meteorites'
    elif row['class_tier_2'] == 'Ungrouped Iron Meteorites':
        return 'Iron Meteorites'
    else:
        return 'Ungrouped Meteorites'
    
meteorClassDataTier1 = meteorClassDataTier2
meteorClassDataTier1['class_tier_1'] = meteorClassDataTier1.apply(class_tier_1, axis=1)
meteorClassDataTier1
```

```python
cols = meteorClassDataTier1.columns.tolist()
print(cols)
cols = cols[3:4] + cols[2:3] + cols[0:1] + cols[1:2]
print(cols)
meteorClassDataFinal = meteorClassDataTier1[cols]
meteorClassDataFinal
```
