```python
import pandas as pd
```

```python
meteorData = pd.read_csv("Meteorite_Landings.csv")
meteorData.head()
```

```python
meteorDataClean = meteorData.drop(columns=["States", "Counties", "fall", "nametype", "id"]).dropna()
meteorDataClean.head()
```

```python
meteorDataClean['year'] = meteorDataClean['year'].str.split(" ").str[0]
```

```python
meteorDataClean['year'] = meteorDataClean['year'].str.split("/").str[-1]
```

```python
meteorDataClean = meteorDataClean.reset_index()
meteorDataClean = meteorDataClean.rename(columns={"index": "id", "mass (g)": "mass_grams"})
meteorDataClean = meteorDataClean.set_index("id")
meteorDataClean = meteorDataClean.loc[meteorDataClean['year'] != "2101"]
```

```python
meteorDataClean.head()
```

```python
meteorDataClean.to_csv("Clean_Meteor_Data.csv")
```

```python
meteorDataClean = meteorDataClean.sort_values("year", ascending=False)
meteorDataClean = meteorDataClean.loc[meteorDataClean["reclat"] != 0]
meteorDataClean.head()
```

```python
meteorDataClean.to_json("Clean_Meteor_Data.json")
```
